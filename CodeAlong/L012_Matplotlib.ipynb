# Matplotlib

```python
import matplotlib.pyplot as plt

x = list(range(-10, 11))
y = [value**2 for value in x]
y2 = [value**3 for value in x]

print(f"{x = }")
print(f"{y = }")

plt.plot(x, y, label="x**2")
plt.plot(x, y2, label="x**3")
plt.legend()
plt.grid()
```

```python
x = [0, -100, 25, 65, -323]
y = [0, 3, 7, 3, 9]

plt.title("My Graph")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.axis([-50, 80, 2, 8])
plt.plot(x, y, color="red", linestyle="dashed", marker="o")
```

## Scatter

```python
x = list(range(-10, 11))
y = [value**2 for value in x]

plt.scatter(x, y, marker="x")
```

## Bars

```python
labels = ["Male", "Female"]
values = [15, 23]

plt.bar(labels, values)
```

## Pie Charts

```python
plt.pie(values, labels=labels)
```

## Sub plots

```python
x = list(range(-10, 11))

y = [value**2 for value in x]
y2 = [value**3 for value in x]

fig, axs = plt.subplots(1, 2)
axs[0].plot(x, y)
axs[1].plot(x, y2)

fig.set_facecolor("red")
axs[0].set_facecolor("green")
axs[1].set_facecolor("blue")
```
